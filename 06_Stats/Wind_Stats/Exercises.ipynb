### Step 1. Import the necessary libraries

```python
import pandas as pd

url = "https://raw.githubusercontent.com/ratul-sraj/pandas_excercises/refs/heads/master/06_Stats/Wind_Stats/wind.data"

# Read without headers
data = pd.read_csv(url, sep=r"\s+", header=None)

# Make first row the header
data.columns = data.iloc[0]
data = data.drop(index=0).reset_index(drop=True)

print(data.head())
```

```python
data.head()
```

```python
data["Yr"] = pd.to_numeric(data["Yr"], errors="coerce")
data["Yr"] = 1900 + data["Yr"]
```

```python
data["Mo"] = pd.to_numeric(data["Mo"], errors="coerce")
data["Dy"] = pd.to_numeric(data["Dy"], errors="coerce")

data["date"] = pd.to_datetime(dict(year=data["Yr"], month=data["Mo"], day=data["Dy"]), errors="coerce")
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
data.shape
```

```python
# Convert all columns (except the index) to numeric
data = data.apply(pd.to_numeric, errors="coerce")

# Now compute row-wise mean
data["mean_all_locations"] = data.mean(axis=1, skipna=True)
data["mean_all_locations"].mean()
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days 

#### A different set of numbers for each location.

```python
stats = {"min": [], "max": [], "mean": [], "std": []}
for col in data.columns:
    stats["min"].append(data.loc[:,col].min())
    stats["max"].append(data.loc[:,col].max())
    stats["mean"].append(data.loc[:,col].mean())
    stats["std"].append(data.loc[:,col].std())

summary_df = pd.DataFrame(stats, index=data.columns)
summary_df    
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python

# Calculate row-wise stats (axis=1 → across columns)
day_stats = pd.DataFrame({
    "min": data.min(axis=1, skipna=True),
    "max": data.max(axis=1, skipna=True),
    "mean": data.mean(axis=1, skipna=True),
    "std": data.std(axis=1, skipna=True)
})
day_stats.head()
```
